```python
total_seats=200
balcony_seat=50
stand_seat=50
balcony_seat_price=500
stand_seat_price=200
total_amount=0
seats_booked=0

while seats_booked<total_seats:
    print("welcome to the Indian Hockey Tournament 2023")
    print("total seats available: ",total_seats-seats_booked)
    print("balcony seats available: ", balcony_seat)
    print("stand seats available: ", stand_seat)
    print("""choose seat type:
        A.for balcony seat
        B. for stand seat""")
    choice=input("enter your choice or press enter to exit")
    
    if choice =="no":
        print("Thank you for interest!")
        break
    
    if choice == "A":
        seat_type = "balcony"
        available_seats = balcony_seat
        seat_price = balcony_seat_price
        
    elif choice =="B":
        seat_type = "stand_seat"
        available_seats = stand_seat
        seat_price = stand_seat_price
        
    else:
        print("invalid option...please print valid option!")
        continue
        
    num_seats = int(input("How many {} seats do you want to book? ".format(seat_type)))
    
    if num_seats > available_seats:
        print("Sorry, only", available_seats, seat_type, "seats are available. Please try again.")
        continue
        
    amount = num_seats * seat_price
    total_amount += amount
    seats_booked += num_seats
    
    if seat_type == "balcony":
        balcony_seat -= num_seats
    else:
        stand_seat -= num_seats
        
    print("Ticket Details:")
    name = input('Enter your name: ')
    phone_number = input('Enter your phone number: ')
    print("Name:", name)
    print("Phone Number:", phone_number)
    print("Seat Type:", seat_type)
    print("Number of seats:", num_seats)
    print("Amount:", amount)
    print("Thank you for booking your tickets!")
    print("=" * 30)

print("All seats are booked. Thank you for your interest!")



```
