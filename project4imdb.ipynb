```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.read_csv(r'C:\Users\PRAGNYAN DASH\Downloads\archive\movies.csv')
 #plot a graph to show the distribution of genre in the top 250 movies 
genre_counts = data['genre'].value_counts()
plt.figure(figsize=(14, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='viridis')
plt.title("Genre Distribution in Top 250 Movies")
plt.xlabel("genre")
plt.ylabel("Count")
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()

data=data.replace(',','',regex=True)

data['imbd_votes'] = data['imbd_votes'].astype(int)

```

```python
#find out which movie has the maximum number of votes and which genre it belongs to and its duration.
max_votes_movie = data[data['imbd_votes'] == data['imbd_votes'].max()]
max_votes_movie_details = max_votes_movie[['title', 'genre', 'duration', 'imbd_votes']]

print("Movie with the maximum number of votes:")
print(max_votes_movie_details)
```

```python
#find out which movie has the minimum number of votes and which genre it belongs to and its duration.
min_votes_movie = data[data['imbd_votes'] == data['imbd_votes'].min()]
min_votes_movie_details = min_votes_movie[['title', 'genre', 'duration', 'imbd_votes']]

print("Movie with the minimum number of votes:")
print(min_votes_movie_details)
```

```python
# Find the movies with the maximum number of votes for each genre
max_votes_per_genre = data.groupby('genre')['imbd_votes'].max()

max_votes_movies = data[data.groupby('genre')['imbd_votes'].transform(max) == data['imbd_votes']]

for genre in max_votes_per_genre.index:
    genre_movies = max_votes_movies[max_votes_movies['genre'] == genre]
    print(f"Movies of {genre} genre with the maximum number of votes:")
    print(genre_movies[['title', 'genre' , 'imbd_votes']])
    print()
```
